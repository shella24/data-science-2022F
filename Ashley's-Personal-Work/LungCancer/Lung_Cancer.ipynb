```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pandas_profiling import ProfileReport
LC = pd.read_csv('survey lung cancer.csv')
```

```python
LC
```

#### 2 = yes 1 = no

## cleaning data

```python
for col in LC.columns:
    pct_missing = np.mean(LC[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
LC = LC.replace([1],0)
LC = LC.replace([2],1)
```

```python
LC['LUNG_CANCER'] = LC['LUNG_CANCER'].replace(['YES'],1)
LC['LUNG_CANCER'] = LC['LUNG_CANCER'].replace(['NO'],0)
```

```python
LC.rename(columns = {"ALCOHOL CONSUMING":"ALCOHOL_CONSUMING", "CHRONIC DISEASE":"CHRONIC_DISEASE","SHORTNESS OF BREATH":"SHORTNESS_OF_BREATH", "SWALLOWING DIFFICULTY":"SWALLOWING_DIFFICULTY","CHEST PAIN":"CHEST_PAIN"}, inplace = True)

```
