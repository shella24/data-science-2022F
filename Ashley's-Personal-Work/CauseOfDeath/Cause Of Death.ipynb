# Cleaning the Data

```python
for col in cod.columns:
    pct_missing = np.mean(cod[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
cod.rename(columns = {"Alzheimer's Disease and Other Dementias":"Alzheimer's Disease", 'Cirrhosis and Other Chronic Liver Diseases':'Chronic Liver Diseases','Fire, Heat, and Hot Substances':'Hot Substances', 'Country/Territory':'Country'}, inplace = True)
```
